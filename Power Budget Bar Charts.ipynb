```python
%pylab notebook
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
##############################Initialise#####################################################
nrow = 1
ncol = 1
xsize = 12
ysize = 6

#set global plotting parameters
linew = 5.0 #line width
#msize = 14 #marker size
axew = 1.0 #axes width

numpad = 10

#set global tick parameters
majw = axew #major tick width
majl = 10.0 #major tick length
minw = axew #minor tick width
minl = 5.0 #minor tick length

#set global font sizes
axefont = 15 #axis label font size
numsize = 15 #axis number font size
legfont = 15 #legend font size
labelfont=20

#############################################################################################
font = {'family': 'serif', 'sans-serif' : 'helvetical', 'size' :20}
plt.rc('font', **font)

labels = ['No Dep.', 'Half Dep.', 'Full Dep.']
maxs = [4.34, 4.71, 5.19] 
means = [2.04, 2.80, 3.91]
mins = [0.68, 2.05, 3.40]
modes2U = [0.02, 0.16, 2.94, 3.43, 4.30, 4.84, 4.90, 5.35]

x = np.arange(len(labels))  # the label locations
width = 0.25  # the width of the bars

fig, ax = plt.subplots(nrows=nrow, ncols=ncol, figsize=(xsize,ysize))
rectsmaxs = ax.bar(x - width, maxs, width, label='Max', color = 'k')
rectsmeans = ax.bar(x, means, width, label='Mean', color = 'dimgrey')
rectsmins = ax.bar(x + width, mins, width, label='Min', color = 'darkgrey')
xdata = linspace(-1,4,10)
plots = ax.plot(xdata,modes2U[0]*linspace(1,1,10), '--', color = 'maroon', label = 'Critical', lw = 2.5)
plots = ax.plot(xdata,modes2U[1]*linspace(1,1,10), color = 'red', label = 'Danger', lw = 2.5)
plots = ax.plot(xdata,modes2U[2]*linspace(1,1,10), '--', color = 'darkorange', label = 'Safe', lw = 2.5)
plots = ax.plot(xdata,modes2U[3]*linspace(1,1,10), color = 'yellow', label = 'Standby', lw = 2.5)
plots = ax.plot(xdata,modes2U[4]*linspace(1,1,10), '--', color = 'deepskyblue', label = 'Detumbling', lw = 2.5)
plots = ax.plot(xdata,modes2U[5]*linspace(1,1,10), '--', color = 'lightsteelblue', label = 'DFGM', lw = 2.5)
plots = ax.plot(xdata,modes2U[6]*linspace(1,1,10), '--', color = 'lightskyblue', label = 'Payload', lw = 2.5)
plots = ax.plot(xdata,modes2U[7]*linspace(1,1,10), color = 'springgreen', label = 'Mixed Science', lw = 2.5)

ax.tick_params(axis='x',which='major',width=majw,length=majl,labelsize=numsize,zorder=1,direction='in',pad=numpad,top='off')
ax.tick_params(axis='x',which='minor',width=minw,length=minl,labelsize=numsize,zorder=1,direction='in',pad=numpad,top='off')
ax.tick_params(axis='y',which='major',width=majw,length=majl,labelsize=numsize,zorder=1,direction='in',pad=numpad,right='off')
ax.tick_params(axis='y',which='minor',width=minw,length=minl,labelsize=numsize,zorder=1,direction='in',pad=numpad,right='off')


# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel(r'$P$ (W)', fontsize = labelfont)
ax.set_xticks(x)
ax.set_xticklabels(labels)
leg = ax.legend(fontsize = legfont,loc='upper right',framealpha = 1)


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize = legfont)


autolabel(rectsmaxs)
autolabel(rectsmeans)
autolabel(rectsmins)

#fig.tight_layout()

ax.set_ylim(0,6)
ax.set_xlim(-0.5,3.5)

plt.show()
```
